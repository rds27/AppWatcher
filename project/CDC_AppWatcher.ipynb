```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm

ios_df = pd.read_csv('763K_plus_IOS_Apps_Info.csv')
ios_df.head()
```

```python
ios_df.drop(ios_df.index[(ios_df["Total_Average_Rating"] == -1)],axis=0,inplace=True)
```

```python
ios_df.drop(ios_df.index[(ios_df["Price_USD"] == -1)],axis=0,inplace=True)
```

```python
ios_df.describe()
```

```python
cols = ios_df.corr().nlargest(4, 'Total_Average_Rating')['Total_Average_Rating'].index
cm = ios_df[cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(cm, annot=True, cmap = 'Blues')
```
